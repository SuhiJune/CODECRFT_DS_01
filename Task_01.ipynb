```python
#IMPORTING REQUIRED LIBRARIES
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Loading dataset
df = pd.read_csv('population_data.csv')

#DATA CLEANING
print(df.isnull().sum())

#BAR CHART FOR GENDER DISTRIBUTION
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', hue='Gender', data=df, palette='Set2')
plt.title('Gender Distribution in the Population')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

#HISTOGRAM FOR GENDER DISTRIBUTION
plt.figure(figsize=(8, 6))
sns.histplot(df['Gender'], bins=2, kde=True, color='skyblue')
plt.title('Gender Distribution in the Population')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
